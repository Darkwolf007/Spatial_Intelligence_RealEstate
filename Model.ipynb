```python
import pandas as pd
import re
import unicodedata

# Function to remove accents
def remove_accents(text):
    return ''.join(c for c in unicodedata.normalize('NFKD', text) if not unicodedata.combining(c))

# Function to clean and normalize text
def clean_text(text):
    if pd.isnull(text):
        return ''
    text = text.lower().strip()  # Lowercase and strip whitespace
    text = re.sub(r'[^\w\s]', '', text)  # Remove special characters
    text = remove_accents(text)  # Remove accents
    text = ' '.join(sorted(text.split()))  # Sort tokens
    return text

# Step 1: Reset index of zoneDf and rename it for clarity
zoneDf_indexed = zoneDf.reset_index().rename(columns={'index': 'zone_index'})

# Step 2: Apply cleaning to both columns
df['area_name_en_clean'] = df['area_name_en'].apply(clean_text)
zoneDf_indexed['CNAME_E_clean'] = zoneDf_indexed['CNAME_E'].apply(clean_text)

# Step 3: Merge using cleaned columns
merged_df = pd.merge(
    df[['area_name_en', 'area_name_en_clean']],
    zoneDf_indexed[['CNAME_E_clean', 'zone_index']],
    left_on='area_name_en_clean',
    right_on='CNAME_E_clean',
    how='left'
)

# Step 4: Create the final DataFrame
new_df = merged_df[['area_name_en', 'zone_index']]

# Step 5: Display the result
display(new_df.head())
```

Adding home typology based on relative home area

```python
# Calculate the min and max procedure_area for each rooms_en category (excluding NaNs in rooms_en)
room_area_ranges = new_df_cleaned.dropna(subset=['rooms_en']).groupby('rooms_en')['procedure_area'].agg(['min', 'max'])

# Define a function to fill missing rooms_en based on procedure_area and calculated ranges
def fill_rooms_by_area_range(row):
    if pd.isna(row['rooms_en']):
        area = row['procedure_area']
        if pd.isna(area):
            return np.nan # Keep as NaN if procedure_area is also NaN
        else:
            for room_type, ranges in room_area_ranges.iterrows():
                if ranges['min'] <= area <= ranges['max']:
                    return room_type
            return np.nan # Return NaN if the area doesn't fall into any defined range
    else:
        return row['rooms_en'] # Keep existing rooms_en value

# Apply the function to fill missing rooms_en values
new_df_cleaned['rooms_en'] = new_df_cleaned.apply(fill_rooms_by_area_range, axis=1)

# Display the head of the updated dataframe
display(new_df_cleaned.head())
```

Smoothing the neighbours for anomalies

```python

year_cols = [col for col in pivot_df.columns if str(col).isdigit()]

smoothed_df = pivot_df.copy()

for (area, room), group in pivot_df.groupby(["area_name_en", "rooms_en"]):
    # Check if there is any price data for the current group
    if not group[year_cols].dropna(axis=1, how='all').empty:
        prices = group[year_cols].iloc[0]  # Select the row of prices for the group

        # rolling median (3 years, centered)
        rolling_med = prices.rolling(window=3, center=True, min_periods=1).median()

        # define anomaly: 3x higher or 3x lower
        mask = (prices > rolling_med * 3) | (prices < rolling_med / 3)

        # replace anomalies
        smoothed_df.loc[group.index, year_cols] = prices.where(~mask, rolling_med)
    else:
        # If no price data, keep the row as is (likely all NaNs in year_cols)
        smoothed_df.loc[group.index, year_cols] = group[year_cols]
```

```python
import pandas as pd
import numpy as np

def fill_missing_only(merged_df):
    """
    ✅ Fills only empty (NaN) cells in smoothed_df by horizontal interpolation.
    ✅ Adds normalized weighted_score column (0–1).
    🚫 Does NOT modify any other existing values.
    """

    df = merged_df.copy()

    # ------------------------------------------------------------
    # Detect year-like columns automatically (e.g., 2000–2030)
    # ------------------------------------------------------------
    year_cols = [c for c in df.columns if str(c).replace('.', '', 1).isdigit()]
    df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')

    # ------------------------------------------------------------
    # STEP 1: Fill only NaN cells (keep existing values)
    # ------------------------------------------------------------
    df[year_cols] = df[year_cols].interpolate(
        axis=1,
        limit_direction='both',
        inplace=False
    )

    # ------------------------------------------------------------
    # STEP 2: Add normalized weighted_score (0–1)
    # ------------------------------------------------------------
    # if weighted_score exists → normalize
    if 'weighted_score' in df.columns:
        ws = pd.to_numeric(df['weighted_score'], errors='coerce')
        df['weighted_score'] = (ws - ws.min()) / (ws.max() - ws.min())
    else:
        # else compute a placeholder score (mean of all yearly values)
        df['weighted_score'] = (
            df[year_cols].mean(axis=1) - df[year_cols].mean(axis=1).min()
        ) / (df[year_cols].mean(axis=1).max() - df[year_cols].mean(axis=1).min())

    print("✅ Filled only missing (NaN) cells and added normalized weighted_score.")
    return df


# ==========================================================
# USE IT DIRECTLY
# ==========================================================
final_df = fill_missing_only(merged_df)

print("✅ final_df ready in memory (only NaNs filled, weighted_score added).")
```

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

# ---------- CONFIG ----------
future_years = [2026, 2027, 2028, 2029, 2030]
model_path = "area_models.pkl"

# ---------- STEP 1: PREPARE DATA ----------
def prepare_data(final_df):
    df = final_df.copy()

    # Identify numeric year columns
    year_cols = [c for c in df.columns if str(c).replace('.', '', 1).isdigit()]

    # Convert to long format
    df_long = df.melt(
        id_vars=["area_name_en", "zone_index", "rooms_en", "weighted_score"],
        value_vars=year_cols,
        var_name="year",
        value_name="price"
    )
    df_long["year"] = df_long["year"].astype(float)
    df_long = df_long.dropna(subset=["price"])

    # Add normalized year for temporal awareness
    df_long["year_norm"] = (
        df_long["year"] - df_long["year"].min()
    ) / (df_long["year"].max() - df_long["year"].min())

    return df_long


# ---------- STEP 2: TRAIN MODELS ----------
def train_all_models(final_df):
    df_long = prepare_data(final_df)
    models = {}
    results = []

    for (area, room), group in df_long.groupby(["area_name_en", "rooms_en"]):
        if group.shape[0] < 5:
            continue  # skip small time series

        X = group[["year_norm", "weighted_score"]]
        y = group["price"]

        # Random Forest for non-linear behavior
        rf = RandomForestRegressor(n_estimators=300, random_state=42)
        rf.fit(X, y)

        # Linear Regression for global trend
        lin = LinearRegression()
        lin.fit(group[["year_norm"]], y)

        models[(area, room)] = {
            "rf": rf,
            "lin": lin,
            "year_min": group["year"].min(),
            "year_max": group["year"].max()
        }

        # Forecast next 5 years
        year_norm_future = (np.array(future_years) - group["year"].min()) / (
            group["year"].max() - group["year"].min()
        )
        weighted_score = group["weighted_score"].iloc[-1]
        X_future = pd.DataFrame({
            "year_norm": year_norm_future,
            "weighted_score": [weighted_score] * len(future_years)
        })

        rf_pred = rf.predict(X_future)
        lin_pred = lin.predict(X_future[["year_norm"]])

        # Blend predictions
        y_pred = 0.8 * rf_pred + 0.2 * lin_pred

        # Add amenity-based growth multiplier
        growth_rate = 0.015 + (0.035 * weighted_score)  # 1.5–5% annual growth
        for i in range(1, len(future_years)):
            y_pred[i] = y_pred[i-1] * (1 + growth_rate)

        temp = pd.DataFrame({
            "area_name_en": area,
            "rooms_en": room,
            "year": future_years,
            "predicted_price_aed": y_pred
        })
        results.append(temp)

    forecast_df = pd.concat(results, ignore_index=True)

    # Save all models
    with open(model_path, "wb") as f:
        pickle.dump(models, f)

    print(f"✅ Trained {len(models)} time-aware models and saved to '{model_path}'")
    return forecast_df, models


# ---------- STEP 3: FORECAST WITH AMENITIES ----------
def forecast_with_amenities(final_df, area, room, added_amenities=None):
    with open(model_path, "rb") as f:
        models = pickle.load(f)

    key = (area, room)
    if key not in models:
        raise ValueError(f"Model for {key} not found.")

    model_pack = models[key]
    rf, lin = model_pack["rf"], model_pack["lin"]
    year_min, year_max = model_pack["year_min"], model_pack["year_max"]

    # Lookup base weighted score
    match = final_df[
        (final_df["area_name_en"].str.strip().str.lower() == area.lower()) &
        (final_df["rooms_en"].str.strip().str.lower() == room.lower())
    ]
    if match.empty:
        raise ValueError(f"No data found for {area}, {room}")

    base_score = match["weighted_score"].iloc[0]
    total_add = sum(added_amenities.values()) if added_amenities else 0
    new_score = min(base_score + total_add, 1.0)

    # Prepare future years
    year_norm_future = (np.array(future_years) - year_min) / (year_max - year_min)
    X_future = pd.DataFrame({
        "year_norm": year_norm_future,
        "weighted_score": [new_score] * len(future_years)
    })

    rf_pred = rf.predict(X_future)
    lin_pred = lin.predict(X_future[["year_norm"]])
    y_pred = 0.8 * rf_pred + 0.2 * lin_pred

    # Add growth adjustment again for amenities
    growth_rate = 0.015 + (0.035 * new_score)
    for i in range(1, len(future_years)):
        y_pred[i] = y_pred[i-1] * (1 + growth_rate)

    result = pd.DataFrame({
        "year": future_years,
        "predicted_price_aed": y_pred,
        "weighted_score_used": new_score
    })
    print(f"✅ Forecast generated for {area} ({room}) — weighted_score {new_score:.3f}")
    return result


# ---------- STEP 4: TRAIN MODELS ----------
forecast_df, area_models = train_all_models(final_df)

# ---------- STEP 5: TEST EXAMPLE ----------
test = forecast_with_amenities(
    final_df,
    area="Abu Hail",
    room="2 B/R",
    added_amenities={"metro": 0.05, "school": 0.03}
)
print("\n🏗️ Forecast for Abu Hail 2 B/R with new amenities:")
print(test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

def validate_model(smoothed_df, models):
    df_long = prepare_data(smoothed_df)

    results = []
    for (area, room), group in df_long.groupby(["area_name_en", "rooms_en"]):
        if group.shape[0] < 8:  # need enough years to validate
            continue

        # chronological split
        group = group.sort_values("year")
        split_idx = int(len(group) * 0.8)
        train, test = group.iloc[:split_idx], group.iloc[split_idx:]

        X_train = train[["year_norm", "weighted_score"]]
        y_train = train["price"]

        X_test = test[["year_norm", "weighted_score"]]
        y_test = test["price"]

        # Refit local model for testing
        rf = RandomForestRegressor(n_estimators=300, random_state=42)
        rf.fit(X_train, y_train)

        y_pred = rf.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        results.append({
            "area_name_en": area,
            "rooms_en": room,
            "MAE": mae,
            "R2": r2,
            "Last_Actual": y_test.values[-1],
            "Last_Predicted": y_pred[-1]
        })

    val_df = pd.DataFrame(results)
    print("✅ Validation complete")
    print(val_df.describe()[["MAE", "R2"]])
    return val_df
```

```python
import pandas as pd
import numpy as np
import pickle
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# ---------- STEP 1: PREPARE LONG DATA ----------
def prepare_data(merged_df):
    df = merged_df.copy()

    # Detect numeric year columns automatically
    year_cols = [c for c in df.columns if str(c).replace('.', '', 1).isdigit()]

    # Melt wide → long format
    df_long = df.melt(
        id_vars=["area_name_en", "zone_index", "rooms_en", "weighted_score"],
        value_vars=year_cols,
        var_name="year",
        value_name="price"
    )

    df_long["year"] = df_long["year"].astype(float)
    df_long = df_long.dropna(subset=["price"])
    df_long = df_long[df_long["price"] > 0]
    df_long = df_long[df_long["price"] < 1e7]
    df_long = df_long.sort_values(["area_name_en", "rooms_en", "year"]).reset_index(drop=True)

    # Temporal normalization
    df_long["year_norm"] = (df_long["year"] - df_long["year"].min()) / (df_long["year"].max() - df_long["year"].min())

    # Momentum features
    df_long["price_prev"] = df_long.groupby(["area_name_en", "rooms_en"])["price"].shift(1)
    df_long["price_change"] = df_long["price"] - df_long["price_prev"]
    df_long["pct_change"] = df_long["price_change"] / df_long["price_prev"]
    df_long = df_long.dropna(subset=["price_prev", "pct_change"])

    # Log-transform for stability
    df_long["log_price"] = np.log1p(df_long["price"])
    df_long.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_long = df_long.dropna(subset=["log_price"])

    print(f"✅ Prepared {len(df_long):,} records | price range: {df_long['price'].min():,.0f}–{df_long['price'].max():,.0f}")
    return df_long


# ---------- STEP 2: TRAIN GLOBAL XGBOOST ----------
def train_global_xgb(final_df, model_path="global_xgb_model.pkl"):
    df_long = prepare_data(final_df)

    X = df_long[["year_norm", "weighted_score", "zone_index", "pct_change"]]
    y = df_long["log_price"]

    model = XGBRegressor(
        n_estimators=500,
        learning_rate=0.07,
        max_depth=9,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    model.fit(X, y)

    # Save model
    with open(model_path, "wb") as f:
        pickle.dump(model, f)

    print(f"✅ Model trained and saved to {model_path}")

    # Evaluate
    y_pred = model.predict(X)
    mae = mean_absolute_error(np.expm1(y), np.expm1(y_pred))
    r2 = r2_score(np.expm1(y), np.expm1(y_pred))
    print(f"📊 Metrics — MAE: {mae:,.0f} AED | R²: {r2:.3f}")

    # Feature importance
    importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": model.feature_importances_
    }).sort_values("Importance", ascending=False)
    print("\n🏗️ Feature Importance:\n", importance)

    return model, df_long, importance


# ---------- STEP 3: RUN ----------
model, df_long, importance = train_global_xgb(merged_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def forecast_area_room(final_df, model, area, room, added_amenities=None):
    # Prepare long data
    df_long = prepare_data(final_df)

    # Filter target area and room
    group = df_long[
        (df_long["area_name_en"].str.strip().str.lower() == area.lower()) &
        (df_long["rooms_en"].str.strip().str.lower() == room.lower())
    ].sort_values("year")

    if group.empty:
        print(f"❌ No data found for {area} – {room}")
        print("\nAvailable areas:", list(df_long["a_name_en"].unique())[:10])
        print("Available rooms:", list(df_long["rooms_en"].unique()))
        return None

    # Extract latest feature info
    zone_index = group["zone_index"].iloc[-1]
    weighted_score = group["weighted_score"].iloc[-1]
    total_add = sum(added_amenities.values()) if added_amenities else 0
    new_score = min(weighted_score + total_add, 1.0)
    last_pct = group["pct_change"].iloc[-1]

    # ---- Historical prediction ----
    X_hist = group[["year_norm", "weighted_score", "zone_index", "pct_change"]]
    y_true = group["price"]
    y_pred_hist = np.expm1(model.predict(X_hist))

    # ---- Future forecast (2025–2030) ----
    future_years = np.arange(2025, 2031)
    year_min, year_max = df_long["year"].min(), df_long["year"].max()
    year_norm_future = (future_years - year_min) / (year_max - year_min)

    X_future = pd.DataFrame({
        "year_norm": year_norm_future,
        "weighted_score": [weighted_score] * len(future_years),
        "zone_index": [zone_index] * len(future_years),
        "pct_change": [last_pct] * len(future_years)
    })

    y_pred_future = np.expm1(model.predict(X_future))

    # Anchor 2025 forecast to match 2024 prediction
    last_pred_2024 = y_pred_hist[-1]
    scale_factor = last_pred_2024 / y_pred_future[0]
    y_pred_future *= scale_factor

    # Apply smooth compounded growth based on score
    growth_rate = 0.015 + (0.035 * weighted_score)
    for i in range(1, len(y_pred_future)):
        y_pred_future[i] = y_pred_future[i-1] * (1 + growth_rate)

    # Amenity multiplier effect
    amenity_factor = 1 + (total_add * 1.5)
    y_pred_future *= amenity_factor

    # ---- Prepare forecast DataFrame ----
    forecast_df = pd.DataFrame({
        "year": future_years,
        "predicted_price_aed": y_pred_future,
        "weighted_score_used": new_score,
        "amenity_factor": amenity_factor
    })

    # ---- Plot actual, predicted, forecasted ----
    plt.figure(figsize=(10, 6))
    plt.plot(group["year"], y_true, label="Actual (2000–2024)", color="royalblue", marker="o")
    plt.plot(group["year"], y_pred_hist, label="Predicted Fit (2000–2024)", color="orange", linestyle="--", marker="x")
    plt.plot(future_years, y_pred_future, label="Forecast (2025–2030)", color="red", linestyle="-.", marker="^")

    plt.title(f"{area} – {room} | Actual, Predicted, and Forecasted Prices")
    plt.xlabel("Year")
    plt.ylabel("Price (AED)")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()

    # ---- Print summary ----
    print(f"\n🏗️ Forecast Summary for {area} ({room}) — weighted_score {weighted_score:.3f} → {new_score:.3f}")
    print(f"📈 Amenities boost applied: +{total_add:.2f} (x{amenity_factor:.2f})")
    print(forecast_df)
    return forecast_df
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

def generate_predicted_dataset(final_df, model, start_year=2000, end_year=2030):
    """
    Generate a full predicted dataset (wide format) for all area-room combinations.
    Uses your XGBoost model and prepare_data() pipeline.
    """

    print(f"🚀 Generating predicted dataset from {start_year}–{end_year} ...")

    # Step 1: Prepare data
    df_long = prepare_data(final_df)

    # Step 2: Initialize empty list for results
    all_records = []

    # Step 3: Iterate per area and room combination
    for (area, zone, room, w_score), group in tqdm(df_long.groupby(["area_name_en", "zone_index", "rooms_en", "weighted_score"]),
                                                   total=df_long[["area_name_en", "rooms_en"]].drop_duplicates().shape[0],
                                                   desc="Predicting"):
        group = group.sort_values("year")
        X_hist = group[["year_norm", "weighted_score", "zone_index", "pct_change"]]
        y_pred_hist = np.expm1(model.predict(X_hist))

        # Store predictions
        for year, price in zip(group["year"], y_pred_hist):
            all_records.append({
                "area_name_en": area,
                "zone_index": zone,
                "rooms_en": room,
                "weighted_score": w_score,
                "year": year,
                "predicted_price_aed": price
            })

    # Step 4: Build combined long DataFrame
    pred_long_df = pd.DataFrame(all_records)

    # Step 5: Pivot to wide format like final_df
    predicted_dataset_df = pred_long_df.pivot_table(
        index=["area_name_en", "zone_index", "rooms_en", "weighted_score"],
        columns="year",
        values="predicted_price_aed"
    ).reset_index()

    # Step 6: Ensure column format consistency (e.g., 2000.0)
    predicted_dataset_df.columns = [str(col) if isinstance(col, float) else col for col in predicted_dataset_df.columns]

    print(f"✅ Predicted dataset ready: {len(predicted_dataset_df)} rows | Years {start_year}–{end_year}")
    return predicted_dataset_df
```
